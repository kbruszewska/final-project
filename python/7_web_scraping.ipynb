```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.numbeo.com/crime/region_rankings_current.jsp?region=150"
```

```python
response = requests.get(url)
```

```python
response.status_code # 200 status code means OK!
```

```python
response.content
```

```python
soup = BeautifulSoup(response.content, "html.parser")
soup
```

```python
len(soup.select('td.cityOrCountryInIndicesTable'))
```

```python
soup.select(".discreet_link")[1].get_text()
```

```python
cityList = []  


for li in soup.select(".discreet_link"):
    cityList.append(li.get_text())  
cityList.pop(0)
cityList[0].split(",")[0]
```

```python
city = []  # list for work pieces 
country = []   # list for composers
for li in cityList:
    #print(li.get_text().split("-"))
    city.append(li.split(",")[0])
    country.append(li.split(",")[1])
```

```python
cityList_df = pd.DataFrame({"city":city,
                           "country":country
                          })
```

```python
cityList_df.head()
```

```python
cityList_df.to_excel('crime_index.xlsx', index=False)
```
