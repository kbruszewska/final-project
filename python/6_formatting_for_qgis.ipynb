```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('../Notebook/data_clean/clusters_subdistricts.csv')
df
```

```python
def format(zahl):
    # Correcting lor: all unique lor should have 6 characters. 
    newZahl = "0"
    if len(str(zahl)) == 5:
        newZahl = newZahl + str(zahl)
    else: newZahl = str(zahl)

    return newZahl
```

```python
df['lor'] = df['lor'].apply(format)
```

```python
pd.set_option('display.max_columns', None)  # to see all columns
df
```

```python
df["ratio_green"] = ((df["area_objects"]/df["area"])*100).round(2)
```
