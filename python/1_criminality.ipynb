```python
def clean_data(df, year):  
    # Function to clean data
    # year should be a string
    df = df.drop(["Gefährl. und schwere Körper-verletzung", "Diebstahl \nan/aus Kfz", "Branddelikte \n-insgesamt-", "Brand-\nstiftung",\
             "Sach-beschädigung -insgesamt-","Kieztaten", "Diebstahl von Kraftwagen", "Straßenraub,\nHandtaschen-raub"], axis=1)
    
    df["hard_crime"] = df["Körper-verletzungen \n-insgesamt-"] + df["Freiheits-beraubung, Nötigung,\nBedrohung, Nachstellung"]
    df["theft_other"] = df["Diebstahl \n-insgesamt-"] - df["Fahrrad-\ndiebstahl"]

    df = df.drop(["Körper-verletzungen \n-insgesamt-", "Freiheits-beraubung, Nötigung,\nBedrohung, Nachstellung",\
             "Diebstahl \n-insgesamt-"], axis=1)
    df = df.rename(columns={'Bezeichnung (Bezirksregion)':'subdistrict', 'Bezirk':'district','Straftaten \n-insgesamt-':'crime_total', \
                        'Raub':'robbery', 'Wohnraum-\neinbruch':'bulglary', 'Fahrrad-\ndiebstahl':'theft_bike',\
                        'Sach-beschädigung durch Graffiti':'graffiti', 'Rauschgift-delikte':'drug_offences'})
    
    df["other_crime"] = df["crime_total"] - df["robbery"] - df["theft_bike"] - df["bulglary"] - df["graffiti"] - df["drug_offences"] - \
    df["hard_crime"] - df["theft_other"]
    
    df.columns = df.columns.str.lower()
    
    df = df[['lor','subdistrict', 'district', 'crime_total', 'robbery', 'theft_bike', 'theft_other', 'bulglary',\
        'graffiti', 'drug_offences', 'hard_crime', 'other_crime']]  

    df = df.rename(columns={"crime_total":"crime_total_"+year, "robbery":"robbery_"+year, "theft_bike":"theft_bike_"+year,\
                                "theft_other":"theft_other_"+year, "bulglary":"bulglary_"+year, "graffiti":"graffiti_"+year,\
                                "drug_offences":"drug_offences_"+year, "hard_crime":"hard_crime_"+year, "other_crime":"other_crime_"+year})

    return df
```
