#### Density

```python
density = pd.concat([lor, population["pop_2013"], population["pop_2014"], population["pop_2015"], population["pop_2016"],\
                     population["pop_2017"], population["pop_2018"], population["pop_2019"], population["pop_2020"],\
                     population["pop_2021"]], axis=1)
```

```python
density
```

```python
# den_2013 means density in 2013 
# density = population/area [in ha]
```

```python
density["den_2013"] = (density["pop_2013"]/density["area"]).round(2)
density["den_2014"] = (density["pop_2014"]/density["area"]).round(2)
density["den_2015"] = (density["pop_2015"]/density["area"]).round(2)
density["den_2016"] = (density["pop_2016"]/density["area"]).round(2)
density["den_2017"] = (density["pop_2017"]/density["area"]).round(2)
density["den_2018"] = (density["pop_2018"]/density["area"]).round(2)
density["den_2019"] = (density["pop_2019"]/density["area"]).round(2)
density["den_2020"] = (density["pop_2020"]/density["area"]).round(2)
density["den_2021"] = (density["pop_2021"]/density["area"]).round(2)
```

```python
density
```

```python
density = density.drop(["pop_2013", "pop_2014", "pop_2015", "pop_2016", "pop_2017", "pop_2018", "pop_2019", "pop_2020", "pop_2021"], axis=1)
```

```python
density = density.drop(["area"], axis=1)
```

```python
density
```

```python
density.to_excel('density.xlsx', index=False)
```
